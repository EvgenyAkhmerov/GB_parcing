```python
import json
from pymongo import MongoClient
from pprint import pprint
from pymongo.errors import DuplicateKeyError
```

```python
client = MongoClient()
db = client.vacancies_hh
vacancies_collections = db.vacancies
```

```python
def from_json_to_doc(file_name: str):
    """return document from json file"""
    with open(file_name, 'r', encoding='UTF-8') as f:
        documents = json.load(f)
    return documents


def add_vacancies_into_db(collection, documents):
    """function inserts docs into db"""
    for doc in documents:
        if not len(list(collection.find(doc))):
          try: 
              collection.insert_one(doc)
          except DuplicateKeyError:
              pass


def vacancies_finding_by_salary(collection, min_salary, max_salary):
    """print vacancies with required salary"""
    col = collection.find({
        '$and': [
            {'min_salary': {'$gte': min_salary}},
            {'max_salary': {'$lte': max_salary}}
        ]
    })

    for doc in col:
        pprint(doc)

```

```python
documents = from_json_to_doc('hh_ru_parcing_ML.json')
```

```python
add_vacancies_into_db(vacancies_collections, documents)
```

```python
for vac in vacancies_collections.find()[:4]: # выведем первые 4 вакансии из базы данных
    pprint(vac)
```

```python
vacancies_finding_by_salary(vacancies_collections, 100000, 200000)
```
